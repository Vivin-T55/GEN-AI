```python
import gradio as gr
from transformers import pipeline

```

```python
pos_tagger = pipeline("ner", model="dslim/bert-base-NER", aggregation_strategy="simple")
```

```python
def tag_pos(text):
    results = pos_tagger(text)
    tagged_words = [(result['word'], result['entity_group']) for result in results]
    return tagged_words
```

```python
app = gr.Interface(
    fn=tag_pos,
    inputs=gr.Textbox(placeholder="Enter your text here..."),
    outputs=gr.Dataframe(headers=["Word", "Part of Speech"]),
    title='Part of Speech Tagging App',
    description='This app tags parts of speech in the given text.'
)

app.launch()
```
